****DATA CLEANING****

```python
Gender_Mod= test_df["Gender"].mode() [0]
test_df["Gender"]. fillna(Gender_Mod, inplace= True)
```

```python
test_df
```

```python
Mod_Dependents= test_df["Dependents"].mode() [0]
test_df["Dependents"].fillna(Mod_Dependents, inplace= True)
test_df
```

```python
Mod_Self_Employed= test_df["Self_Employed"].mode() [0]
test_df["Self_Employed"].fillna(Mod_Self_Employed, inplace= True)
test_df
```

```python
LoanAmount_Mid= test_df["LoanAmount"].median() 
test_df["LoanAmount"].fillna(LoanAmount_Mid, inplace= True)
test_df
```

```python
Loan_Term_Mid= test_df["Loan_Amount_Term"].median() 
test_df["Loan_Amount_Term"].fillna(Loan_Term_Mid, inplace= True)
test_df
```

```python
Credit_History_Mid= test_df["Credit_History"].median() 
test_df["Credit_History"].fillna(Credit_History_Mid, inplace= True)
test_df
```

```python
#proper plotting
def univariate_cat(col):
    EducationRate = test_df[col].value_counts()
    list_p = EducationRate.values
    total_p= sum(EducationRate.values)
    sns.set_palette("Set2") # for coloring
    fig, ola = plt.subplots(figsize= (10, 6))
    EducationRate.plot(kind= "bar", ax= ola ) # pandas plot
    plt.grid(axis= "y")
    plt.xticks(color= "red", rotation= 30)
    plt.yticks(color= "red", rotation= 30)
    plt.title( f"count of {col}", color= 'red')
    for index, val in enumerate(list_p):
        plt.text(x=index, y= val+0.5, s= f"{round((val/total_p)*100)}%"  )
    plt.show()
```

```python
univariate_cat("Education")
```

```python
univariate_cat("Gender")
```

```python
univariate_cat("Married")
```

```python
univariate_cat("Dependents")
```

```python
univariate_cat("Self_Employed")
```
