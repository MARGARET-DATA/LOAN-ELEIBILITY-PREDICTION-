```python
#proper plotting
def univariate_cat(col):
    Self_Employed_Rate = train_df[col].value_counts()
    list_p = Self_Employed_Rate.values
    total_p= sum(Self_Employed_Rate.values)
    sns.set_palette("Set2") # for coloring
    fig, ola = plt.subplots(figsize= (10, 6))
    Self_Employed_Rate.plot(kind= "bar", ax= ola ) # pandas plot
    plt.grid(axis= "y")
    plt.xticks(color= "red", rotation= 30)
    plt.yticks(color= "red", rotation= 30)
    plt.title( f"count of {col}", color= 'red')
    for index, val in enumerate(list_p):
        plt.text(x=index, y= val+0.5, s= f"{round((val/total_p)*100)}%"  )
    plt.show()
```

```python
def bivariate(col1, col2):
    bi=pd.crosstab(index = train_df[col1], 
        columns= train_df[col2],
        values= train_df[col2],
        aggfunc= "count")
    
    fig, ax= plt.subplots(figsize=(10,6) ) 
    bi.plot(kind= "bar", ax=ax)
    plt.grid(axis= "y")
    plt.xticks(color= "red", rotation= 30)
    plt.yticks(color= "red", rotation= 30)
    plt.title( f"{col1}  by {col2}" )
```

```python
bivariate("Education" , "Gender")
```

```python
bivariate("Married" , "Gender")
```

```python
bivariate( "Self_Employed", "Married")
```
